```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import math 
```

```python
file = uproot.open('outAna.root')
file.keys()
data = file['cutFlowAnalyzerPXBL4PXFL3;1']
data.keys()
events=data['Events;1']
events.keys()
```

```python
genMu0_eta=events.array('genMu0_eta');
genMu1_eta =events.array('genMu1_eta');
genMu2_eta=events.array('genMu2_eta');
genMu3_eta=events.array('genMu3_eta');
selMu0_eta=events.array('selMu0_eta');
selMu1_eta=events.array('selMu1_eta');
selMu2_eta=events.array('selMu2_eta');
selMu3_eta=events.array('selMu3_eta');
genMu0_phi=events.array('genMu0_phi');
genMu1_phi=events.array('genMu1_phi');
genMu2_phi=events.array('genMu2_phi');
genMu3_phi=events.array('genMu3_phi');
selMu0_phi=events.array('selMu0_phi');
selMu1_phi=events.array('selMu1_phi');
selMu2_phi=events.array('selMu2_phi');
selMu3_phi=events.array('selMu3_phi');
```

```python
Genpt0=events.array('genMu0_pT')
```

```python
N=Genpt0.size
```

```python
dRgen0=np.ndarray(shape=(N,4))
min1=np.zeros((N,))
j=0
k=0
l=0
m=0
matches={}
for ii in range(N): 
    dRgen0[ii][0] = math.sqrt(pow(genMu0_eta[ii]-selMu0_eta[ii],2)+ pow(genMu0_phi[ii]-selMu0_phi[ii],2))        
    dRgen0[ii][1] = math.sqrt(pow(genMu0_eta[ii]-selMu1_eta[ii],2)+ pow(genMu0_phi[ii]-selMu1_phi[ii],2))
    dRgen0[ii][2] = math.sqrt(pow(genMu0_eta[ii]-selMu2_eta[ii],2)+ pow(genMu0_phi[ii]-selMu2_phi[ii],2))
    dRgen0[ii][3] = math.sqrt(pow(genMu0_eta[ii]-selMu3_eta[ii],2)+ pow(genMu0_phi[ii]-selMu3_phi[ii],2))
    
    min1[ii]=dRgen0[ii].min()
    if min1[ii]==dRgen0[ii][0] and abs(selMu0_eta[ii]<=2.4) and selMu0_phi[ii]!=-100 and abs(genMu0_eta[ii])<=2.4 and genMu0_phi[ii]!=-100:
        j=j+1
      #  matches[ii,1,1] = dRgen0[ii][0]
    elif min1[ii]==dRgen0[ii][1] and abs(selMu1_eta[ii])<=2.4 and selMu1_phi[ii]!=-100 and abs(genMu0_eta[ii])<=2.4 and genMu0_phi[ii]!=-100:
        k=k+1 
        matches[ii,1,2] = dRgen0[ii][1]   
    elif min1[ii]==dRgen0[ii][2] and abs(selMu2_eta[ii])<=2.4 and selMu2_phi[ii]!=-100 and abs(genMu0_eta[ii])<=2.4 and genMu0_phi[ii]!=-100:
        l=l+1 
        matches[ii,1,3] = dRgen0[ii][2]   
    elif min1[ii]==dRgen0[ii][3] and abs(selMu3_eta[ii])<=2.4 and selMu3_phi[ii]!=-100 and abs(genMu0_eta[ii])<=2.4 and genMu0_phi[ii]!=-100:
        m=m+1 
        matches[ii,1,4] = dRgen0[ii][3]   
print ('1rst gen matched to 1rst reco: ',  j)        
print ('1rst gen matched to 2nd reco: ',k)
print ('1rst gen matched to 3rd reco: ',l)
print ('1rst gen matched to 4th reco: ',m)
```
